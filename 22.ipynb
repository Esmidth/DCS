```python
from itertools import combinations
```

```python
for c in combinations(total_id_list,2):
    tmp_sum = np.zeros_like(results.T[0])
    tmp_sum = tmp_sum+results_ones.T[c[0]]
    tmp_sum = tmp_sum+results_ones.T[c[1]]
    fig = plt.Figure(figsize=(20,15))
    ax = fig.add_subplot(111)
    ax.plot(tmp_sum)
    # plt.figure(figsize=(20,15))
    # plt.plot(tmp_sum)
    fig.savefig('combinations/{}.png'.format(c),dpi=300,pad_inches=0,bbox_inches='tight')
    fig.clf()
    del tmp_sum
    # plt.savefig('combinations/{}.png'.format(c))
```

```python
ratio_result = {}
for c in combinations(total_id_list,2):
    tmp_sum = np.ones_like(results.T[0])
    tmp_sum = tmp_sum * results_ones.T[c[0]]
    tmp_sum = tmp_sum * results_ones.T[c[1]]
    ratio = len(np.where(tmp_sum == 1)[0]) / len(tmp_sum)
    # 两个传感器数据的交集
    ratio_result[c] = ratio
    # plt.plot(tmp_sum)

    print(c,ratio)

    # fig = plt.Figure(figsize=(20,15))
    # ax = fig.add_subplot(111)
    # ax.plot(tmp_sum)
    # plt.figure(figsize=(20,15))
    # plt.plot(tmp_sum)
    # fig.savefig('combinations/{}.png'.format(c),dpi=300,pad_inches=0,bbox_inches='tight')
    # fig.clf()
    # del tmp_sum
    
    # plt.savefig('combinations/{}.png'.format(c))
print(ratio_result)
```
