```python
# test the data correction in results
# 输出{time}_time_length_{}.png图
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
```

```python
with open('readings_timeid.pkl','rb') as f:
    results = pickle.load(f)
    
```

```python
total_id_list = [  3,   5,   7,   8,   9,  10,  11,  12,  13,  14,  15,  17,  18,
        19,  21,  23,  24,  25,  26,  27,  30,  31,  32,  33,  34,  35,
        36,  37,  39,  40,  41,  42,  43,  44,  45,  46,  47,  49,  50,
        51,  53,  54,  55,  56,  57,  59,  60,  61,  62,  63,  64,  65,
        66,  68,  69,  70,  71,  72,  73,  75,  76,  79,  80,  81,  82,
        84,  85,  87,  88,  89,  92,  93,  94,  95,  96,  97,  98,  99,
       100, 103, 104, 105, 106, 107, 109, 111, 121, 122]
       
```

```python
time_length = 200

for num in np.linspace(0,len(results)-time_length*2,200):
    # print(int(i))
    num = int(num)

    tmp_output = np.zeros((time_length,len(total_id_list)))

    for i in range(time_length):
        tmp_output[i] = results[i+num][total_id_list]

    fig = plt.Figure(figsize=(40,15))
    ax = fig.add_subplot(111)
    ax.imshow(tmp_output.T)
    ax.set_title('#{}'.format(num))
    fig.savefig('figures_1_12/200/{}_time_length_{}.png'.format(num,time_length),dpi=300,pad_inches=0,bbox_inches='tight')
    fig.clf()
```
